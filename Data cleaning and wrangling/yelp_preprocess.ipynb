```python
# This file preprocesses the data from the yelp dataset. We will break down each individual dataset:

import pandas as pd
import numpy as np
```

## We will start with the 'businesses' dataset 

```python
res_df = pd.read_json("yelp_academic_dataset_business.json", lines=True)
display(res_df)
```

```python
desired_categories = ['Restaurants', 'Food', 'Coffee', 'Tea']
res_df = res_df[res_df['categories'].str.contains('|'.join(desired_categories), case=False, na=False)]
```

```python
rows_with_null = res_df[res_df.isnull().any(axis=1)]
# rows_with_null
```

```python
null_categories = res_df[res_df['categories'].isnull()]
# null_categories
```

```python
res_df = res_df.dropna(subset=['categories'])
```

```python
res_df.isna().sum()
res_df['attributes'].fillna({}, inplace=True)
res_df.isna().sum()
```

```python
res_df['attributes'] = res_df['attributes'].apply(lambda x: {} if pd.isnull(x) else x)
```

```python
res_df.drop('hours', axis=1, inplace=True)
res_df.isna().sum()
```

```python
res_df.to_csv('yelp_restaurants.csv', index=False)
```

## Now we will move onto the reviews  ##
This dataset is very large and so i'm having trouble reading it into a dataframe...

```python
r_dtypes = {"stars": np.float16, 
            "useful": np.int32, 
            "funny": np.int32,
            "cool": np.int32}

# List to store filtered review chunks
filtered_chunks = []

# Filter reviews based on restaurant business IDs
with open("yelp_academic_dataset_review.json", "r") as f:
    reader = pd.read_json(f, orient="records", lines=True, dtype=r_dtypes, chunksize=1000)

    for chunk in reader:
        # Merge with restaurant dataset to filter reviews for restaurants only
        merged_chunk = pd.merge(chunk, res_df[['business_id']], on='business_id', how='inner')
        
        # Optionally perform additional filtering or drop unnecessary columns from merged_chunk
        
        # Append the filtered chunk to the list
        filtered_chunks.append(merged_chunk)

# Concatenate all filtered chunks into a single DataFrame
filtered_reviews = pd.concat(filtered_chunks, ignore_index=True)
```
